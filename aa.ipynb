```python
#原始图像标记
#cell5
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
from pathlib import Path

plt.close('all')

try:
    get_ipython().run_line_magic('matplotlib', 'tk')
except Exception as e:
    print('Could not run %matplotlib tk:', e)

try:
    plt.switch_backend('TkAgg')
except Exception as e:
    print('Could not switch to TkAgg:', e)

plt.close('all')


def get_case_name_from_csv_path(path):
    path = Path(path)
    parts = path.parts
    if 'exports' in parts:
        export_idx = parts.index('exports')
        if export_idx > 0:
            return parts[export_idx - 1]
    return path.parents[2].name


case_name = get_case_name_from_csv_path(csv_path) if 'csv_path' in globals() else ''
group_gap_threshold = 10.0

raw = df_eye1.sort_values('pupil_timestamp').copy()
raw['time_s'] = raw['pupil_timestamp'] - raw['pupil_timestamp'].iloc[0]
raw = raw.dropna(subset=['time_s', 'ellipse_center_x']).reset_index(drop=True)

curve_x = raw['time_s'].to_numpy(dtype=float)
curve_y = raw['ellipse_center_x'].to_numpy(dtype=float)

clicked_points = []
markers = []
clicked_points_df = pd.DataFrame(columns=['time_s', 'ellipse_center_x', 'raw_index', 'pupil_timestamp', 'group'])
distance_summary_df = pd.DataFrame(columns=[
    'name', 'n_points', 'small_n', 'large_n', 'small_values', 'large_values',
    'small_mean', 'large_mean', 'max_gap', 'distance_calculate', 'status'
])
coord_table = display(clicked_points_df, display_id=True)
distance_table = display(distance_summary_df, display_id=True)

fig, ax = plt.subplots(num='ellipse_center_x_picker', figsize=(12, 5), clear=True)
ax.plot(curve_x, curve_y, linewidth=1)
ax.set_xlabel('time_s')
ax.set_ylabel('ellipse_center_x')
ax.set_title('Left click: snap to curve | Right click: undo | Mouse wheel: zoom')
ax.grid(True, alpha=0.3)


def classify_by_largest_gap(points):
    if len(points) < 2:
        return None

    values = np.array([float(p['ellipse_center_x']) for p in points], dtype=float)
    order = np.argsort(values)
    sorted_values = values[order]
    gaps = np.diff(sorted_values)
    split_pos = int(np.argmax(gaps))
    max_gap = float(gaps[split_pos])

    if max_gap <= group_gap_threshold:
        return {
            'status': f'Need a gap > {group_gap_threshold:g} between small and large groups. Current max_gap={max_gap:.4f}.',
            'max_gap': max_gap,
        }

    small_idx = {int(i) for i in order[:split_pos + 1]}
    large_idx = {int(i) for i in order[split_pos + 1:]}
    small_values = values[list(small_idx)]
    large_values = values[list(large_idx)]
    small_mean = float(np.mean(small_values))
    large_mean = float(np.mean(large_values))

    return {
        'small_idx': small_idx,
        'large_idx': large_idx,
        'small_values': small_values,
        'large_values': large_values,
        'small_mean': small_mean,
        'large_mean': large_mean,
        'max_gap': max_gap,
        'distance_calculate': large_mean - small_mean,
        'status': 'Calculated only; not saved.',
    }


def build_summary(result, status):
    if result is None or 'distance_calculate' not in result:
        groups = [p.get('group', '') for p in clicked_points]
        small_points = [p for p in clicked_points if p.get('group') == 'small']
        large_points = [p for p in clicked_points if p.get('group') == 'large']
        max_gap = np.nan if result is None else result.get('max_gap', np.nan)
        return pd.DataFrame([{
            'name': case_name,
            'n_points': len(clicked_points),
            'small_n': len(small_points),
            'large_n': len(large_points),
            'small_values': ', '.join(f'{p["ellipse_center_x"]:.4f}' for p in small_points),
            'large_values': ', '.join(f'{p["ellipse_center_x"]:.4f}' for p in large_points),
            'small_mean': np.nan,
            'large_mean': np.nan,
            'max_gap': max_gap,
            'distance_calculate': np.nan,
            'status': status,
        }], columns=distance_summary_df.columns)

    return pd.DataFrame([{
        'name': case_name,
        'n_points': len(clicked_points),
        'small_n': len(result['small_values']),
        'large_n': len(result['large_values']),
        'small_values': ', '.join(f'{v:.4f}' for v in sorted(result['small_values'])),
        'large_values': ', '.join(f'{v:.4f}' for v in sorted(result['large_values'])),
        'small_mean': result['small_mean'],
        'large_mean': result['large_mean'],
        'max_gap': result['max_gap'],
        'distance_calculate': result['distance_calculate'],
        'status': status,
    }], columns=distance_summary_df.columns)


def refresh_table(status=None):
    global clicked_points_df, distance_summary_df

    result = classify_by_largest_gap(clicked_points)
    if result is None:
        status = status or 'Need at least two marked points.'
        for point in clicked_points:
            point['group'] = ''
        for marker in markers:
            marker.set_color('red')
    elif 'distance_calculate' not in result:
        status = status or result['status']
        for point in clicked_points:
            point['group'] = ''
        for marker in markers:
            marker.set_color('red')
    else:
        status = status or result['status']
        for i, point in enumerate(clicked_points):
            point['group'] = 'small' if i in result['small_idx'] else 'large'
        for point, marker in zip(clicked_points, markers):
            marker.set_color('tab:blue' if point['group'] == 'small' else 'tab:red')

    clicked_points_df = pd.DataFrame(clicked_points, columns=['time_s', 'ellipse_center_x', 'raw_index', 'pupil_timestamp', 'group'])
    coord_table.update(clicked_points_df)
    distance_summary_df = build_summary(result, status)
    distance_table.update(distance_summary_df)
    return result


def print_current_distance(prefix='Current'):
    result = classify_by_largest_gap(clicked_points)
    if result is None:
        print(f'{prefix}: need at least two marked points.')
        return
    if 'distance_calculate' not in result:
        print(f'{prefix}: {result["status"]}')
        return

    print(
        f'{prefix}: {case_name} distance_calculate = {result["distance_calculate"]:.6f} '
        f'(large_mean={result["large_mean"]:.6f}, small_mean={result["small_mean"]:.6f}, '
        f'large_n={len(result["large_values"])}, small_n={len(result["small_values"])}, max_gap={result["max_gap"]:.6f})'
    )


def nearest_curve_point(event):
    points_px = ax.transData.transform(np.column_stack([curve_x, curve_y]))
    dist2 = (points_px[:, 0] - event.x) ** 2 + (points_px[:, 1] - event.y) ** 2
    idx = int(np.argmin(dist2))
    return idx, float(curve_x[idx]), float(curve_y[idx])


def toolbar_is_active():
    toolbar = getattr(fig.canvas, 'toolbar', None)
    mode = getattr(toolbar, 'mode', '') if toolbar is not None else ''
    return bool(str(mode))


def onclick(event):
    if event.inaxes != ax:
        return

    # Do not mark points while the toolbar Pan/Zoom tool is active.
    if toolbar_is_active():
        return

    if event.button == 3:
        if clicked_points:
            clicked_points.pop()
            markers.pop().remove()
            refresh_table()
            fig.canvas.draw_idle()
        return

    if event.button != 1:
        return

    idx, x, y = nearest_curve_point(event)
    point = {
        'time_s': x,
        'ellipse_center_x': y,
        'raw_index': idx,
        'pupil_timestamp': float(raw.loc[idx, 'pupil_timestamp']) if 'pupil_timestamp' in raw.columns else np.nan,
        'group': '',
    }
    clicked_points.append(point)
    markers.append(ax.scatter([x], [y], color='red', s=40, zorder=5))
    refresh_table()
    fig.canvas.draw_idle()


def onscroll(event):
    if event.inaxes != ax or event.xdata is None or event.ydata is None:
        return
    scale = 0.8 if event.button == 'up' else 1.25
    x0, x1 = ax.get_xlim()
    y0, y1 = ax.get_ylim()
    x, y = event.xdata, event.ydata
    ax.set_xlim(x - (x - x0) * scale, x + (x1 - x) * scale)
    ax.set_ylim(y - (y - y0) * scale, y + (y1 - y) * scale)
    fig.canvas.draw_idle()


def onkey(event):
    if event.key == 'p':
        print_current_distance('Pressed p')


def onclose(event):
    print_current_distance('Final')


refresh_table()
cid_click = fig.canvas.mpl_connect('button_press_event', onclick)
cid_scroll = fig.canvas.mpl_connect('scroll_event', onscroll)
cid_key = fig.canvas.mpl_connect('key_press_event', onkey)
cid_close = fig.canvas.mpl_connect('close_event', onclose)
print(f'Case name: {case_name}')
print(f'Groups are split automatically at the largest ellipse_center_x gap if max_gap > {group_gap_threshold:g}.')
print('Left click snaps to curve; right click undoes; mouse wheel zooms around cursor.')
print('The table updates the calculated distance. Press p or close the figure window to print the current/final value.')
plt.show(block=True)
```

```python
# cell10
# 10帧合并成1帧取平均值后的标点操作
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
from pathlib import Path

plt.close('all')

try:
    get_ipython().run_line_magic('matplotlib', 'tk')
except Exception as e:
    print('Could not run %matplotlib tk:', e)

try:
    plt.switch_backend('TkAgg')
except Exception as e:
    print('Could not switch to TkAgg:', e)

plt.close('all')


def get_case_name_from_csv_path(path):
    path = Path(path)
    parts = path.parts
    if 'exports' in parts:
        export_idx = parts.index('exports')
        if export_idx > 0:
            return parts[export_idx - 1]
    return path.parents[2].name


case_name = get_case_name_from_csv_path(csv_path) if 'csv_path' in globals() else ''
group_gap_threshold = 10.0
compress_n = 10

# 每次运行都从当前 df_eye1 重新计算 10 帧平均，避免复用旧变量
x_raw = df_eye1.sort_values('pupil_timestamp').copy().reset_index().rename(columns={'index': 'original_index'})
x_raw['time_s'] = x_raw['pupil_timestamp'] - x_raw['pupil_timestamp'].iloc[0]
x_raw = x_raw.dropna(subset=['time_s', 'ellipse_center_x']).reset_index(drop=True)
x_raw['compress_group'] = x_raw.index // compress_n

avg = x_raw.groupby('compress_group', as_index=False).agg(
    time_s=('time_s', 'mean'),
    pupil_timestamp=('pupil_timestamp', 'mean'),
    ellipse_center_x_mean=('ellipse_center_x', 'mean'),
    ellipse_center_x_std=('ellipse_center_x', 'std'),
    original_start_index=('original_index', 'first'),
    original_end_index=('original_index', 'last'),
    n_frames=('ellipse_center_x', 'size')
).dropna(subset=['time_s', 'ellipse_center_x_mean']).reset_index(drop=True)

curve_x = avg['time_s'].to_numpy(dtype=float)
curve_y = avg['ellipse_center_x_mean'].to_numpy(dtype=float)

clicked_10_points = []
markers = []
clicked_10_points_df = pd.DataFrame(columns=[
    'time_s', 'ellipse_center_x_mean', 'compressed_index', 'pupil_timestamp',
    'original_start_index', 'original_end_index', 'n_frames', 'group'
])
distance_summary_df = pd.DataFrame(columns=[
    'name', 'n_points', 'small_n', 'large_n', 'small_values', 'large_values',
    'small_mean', 'large_mean', 'max_gap', 'distance_calculate', 'status'
])
coord_table = display(clicked_10_points_df, display_id=True)
distance_table = display(distance_summary_df, display_id=True)

fig, ax = plt.subplots(num='ellipse_center_x_10frame_mean_picker', figsize=(12, 5), clear=True)
ax.plot(curve_x, curve_y, linewidth=1.5, color='tab:orange')
ax.set_xlabel('time_s')
ax.set_ylabel('ellipse_center_x_mean')
ax.set_title('10-frame mean ellipse_center_x | Left click: snap | Right click: undo | Mouse wheel: zoom')
ax.grid(True, alpha=0.3)


def classify_by_largest_gap(points):
    if len(points) < 2:
        return None

    values = np.array([float(p['ellipse_center_x_mean']) for p in points], dtype=float)
    order = np.argsort(values)
    sorted_values = values[order]
    gaps = np.diff(sorted_values)
    split_pos = int(np.argmax(gaps))
    max_gap = float(gaps[split_pos])

    if max_gap <= group_gap_threshold:
        return {
            'status': f'Need a gap > {group_gap_threshold:g} between small and large groups. Current max_gap={max_gap:.4f}.',
            'max_gap': max_gap,
        }

    small_idx = {int(i) for i in order[:split_pos + 1]}
    large_idx = {int(i) for i in order[split_pos + 1:]}
    small_values = values[list(small_idx)]
    large_values = values[list(large_idx)]
    small_mean = float(np.mean(small_values))
    large_mean = float(np.mean(large_values))

    return {
        'small_idx': small_idx,
        'large_idx': large_idx,
        'small_values': small_values,
        'large_values': large_values,
        'small_mean': small_mean,
        'large_mean': large_mean,
        'max_gap': max_gap,
        'distance_calculate': large_mean - small_mean,
        'status': 'Calculated only; not saved.',
    }


def build_summary(result, status):
    if result is None or 'distance_calculate' not in result:
        small_points = [p for p in clicked_10_points if p.get('group') == 'small']
        large_points = [p for p in clicked_10_points if p.get('group') == 'large']
        max_gap = np.nan if result is None else result.get('max_gap', np.nan)
        return pd.DataFrame([{
            'name': case_name,
            'n_points': len(clicked_10_points),
            'small_n': len(small_points),
            'large_n': len(large_points),
            'small_values': ', '.join(f'{p["ellipse_center_x_mean"]:.4f}' for p in small_points),
            'large_values': ', '.join(f'{p["ellipse_center_x_mean"]:.4f}' for p in large_points),
            'small_mean': np.nan,
            'large_mean': np.nan,
            'max_gap': max_gap,
            'distance_calculate': np.nan,
            'status': status,
        }], columns=distance_summary_df.columns)

    return pd.DataFrame([{
        'name': case_name,
        'n_points': len(clicked_10_points),
        'small_n': len(result['small_values']),
        'large_n': len(result['large_values']),
        'small_values': ', '.join(f'{v:.4f}' for v in sorted(result['small_values'])),
        'large_values': ', '.join(f'{v:.4f}' for v in sorted(result['large_values'])),
        'small_mean': result['small_mean'],
        'large_mean': result['large_mean'],
        'max_gap': result['max_gap'],
        'distance_calculate': result['distance_calculate'],
        'status': status,
    }], columns=distance_summary_df.columns)


def refresh_table(status=None):
    global clicked_10_points_df, distance_summary_df

    result = classify_by_largest_gap(clicked_10_points)
    if result is None:
        status = status or 'Need at least two marked points.'
        for point in clicked_10_points:
            point['group'] = ''
        for marker in markers:
            marker.set_color('red')
    elif 'distance_calculate' not in result:
        status = status or result['status']
        for point in clicked_10_points:
            point['group'] = ''
        for marker in markers:
            marker.set_color('red')
    else:
        status = status or result['status']
        for i, point in enumerate(clicked_10_points):
            point['group'] = 'small' if i in result['small_idx'] else 'large'
        for point, marker in zip(clicked_10_points, markers):
            marker.set_color('tab:blue' if point['group'] == 'small' else 'tab:red')

    clicked_10_points_df = pd.DataFrame(clicked_10_points, columns=[
        'time_s', 'ellipse_center_x_mean', 'compressed_index', 'pupil_timestamp',
        'original_start_index', 'original_end_index', 'n_frames', 'group'
    ])
    coord_table.update(clicked_10_points_df)
    distance_summary_df = build_summary(result, status)
    distance_table.update(distance_summary_df)
    return result


def print_current_distance(prefix='Current'):
    result = classify_by_largest_gap(clicked_10_points)
    if result is None:
        print(prefix + ': need at least two marked points.')
        return
    if 'distance_calculate' not in result:
        print(prefix + ': ' + result['status'])
        return

    print(
        f'{prefix}: {case_name} distance_calculate = {result["distance_calculate"]:.6f} '
        f'(large_mean={result["large_mean"]:.6f}, small_mean={result["small_mean"]:.6f}, '
        f'large_n={len(result["large_values"])}, small_n={len(result["small_values"])}, max_gap={result["max_gap"]:.6f})'
    )


def nearest_curve_point(event):
    points_px = ax.transData.transform(np.column_stack([curve_x, curve_y]))
    dist2 = (points_px[:, 0] - event.x) ** 2 + (points_px[:, 1] - event.y) ** 2
    idx = int(np.argmin(dist2))
    return idx, float(curve_x[idx]), float(curve_y[idx])


def toolbar_is_active():
    toolbar = getattr(fig.canvas, 'toolbar', None)
    mode = getattr(toolbar, 'mode', '') if toolbar is not None else ''
    return bool(str(mode))


def onclick(event):
    if event.inaxes != ax:
        return

    if toolbar_is_active():
        return

    if event.button == 3:
        if clicked_10_points:
            clicked_10_points.pop()
            markers.pop().remove()
            refresh_table()
            fig.canvas.draw_idle()
        return

    if event.button != 1:
        return

    idx, x, y = nearest_curve_point(event)
    row = avg.loc[idx]
    point = {
        'time_s': x,
        'ellipse_center_x_mean': y,
        'compressed_index': idx,
        'pupil_timestamp': float(row['pupil_timestamp']) if 'pupil_timestamp' in avg.columns else np.nan,
        'original_start_index': int(row['original_start_index']) if 'original_start_index' in avg.columns else np.nan,
        'original_end_index': int(row['original_end_index']) if 'original_end_index' in avg.columns else np.nan,
        'n_frames': int(row['n_frames']) if 'n_frames' in avg.columns else np.nan,
        'group': '',
    }
    clicked_10_points.append(point)
    markers.append(ax.scatter([x], [y], color='red', s=45, zorder=5))
    refresh_table()
    fig.canvas.draw_idle()


def onscroll(event):
    if event.inaxes != ax or event.xdata is None or event.ydata is None:
        return
    scale = 0.8 if event.button == 'up' else 1.25
    x0, x1 = ax.get_xlim()
    y0, y1 = ax.get_ylim()
    x, y = event.xdata, event.ydata
    ax.set_xlim(x - (x - x0) * scale, x + (x1 - x) * scale)
    ax.set_ylim(y - (y - y0) * scale, y + (y1 - y) * scale)
    fig.canvas.draw_idle()


def onkey(event):
    if event.key == 'p':
        print_current_distance('Pressed p')


def onclose(event):
    print_current_distance('Final')


refresh_table()
cid_click = fig.canvas.mpl_connect('button_press_event', onclick)
cid_scroll = fig.canvas.mpl_connect('scroll_event', onscroll)
cid_key = fig.canvas.mpl_connect('key_press_event', onkey)
cid_close = fig.canvas.mpl_connect('close_event', onclose)
print(f'Case name: {case_name}')
print(f'10-frame mean points: {len(avg)} from raw points: {len(x_raw)}')
print(f'Groups are split automatically at the largest ellipse_center_x_mean gap if max_gap > {group_gap_threshold:g}.')
print('Left click snaps to the 10-frame mean curve; right click undoes; mouse wheel zooms around cursor.')
print('The table updates the calculated distance. Press p or close the figure window to print the current/final value.')
plt.show(block=True)
```
